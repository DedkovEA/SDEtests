```julia
using DifferentialEquations
using Plots; plotly()
```

```julia
A = [[-0.1 -2.0]
     [ 2.0 -0.1]];

B_1 = 10*[[0. -0.01]
          [ 0.01 -0.01]];
B_2 = 10*[[0. -0.02]
          [ -0.02 0.03]];
#B_1 = zeros((2,2));
```

```julia
tspan = (0., 10.)
dt = 10. / 1000
x_0 = [3.0; 0.0]
f_analytic(u₀,p,t,W) = exp(A*t + B_1*W[1] + B_2*W[2])*u₀;
```

```julia
function f(du,u,p,t)
    du[:] = (A + 0.5*(B_1 + B_2)) * reshape(u, (2,1))
end;

function G(du,u,p,t)
    du[:,:] = [B_1*reshape(u, (2,1)) B_2*reshape(u, (2,1))]
end;

ff = SDEFunction(f,G,analytic=f_analytic)

prob = SDEProblem(ff,G,x_0,(0., 10.),(0.0,1.0), noise_rate_prototype=zeros(2,2))
```

```julia
dt=10. /1000
sol1 = solve(prob, EM(), dt=dt);
```

```julia
sol2 = solve(prob, SRA(), dt=dt);
```

```julia
plot(sol1,plot_analytic=true)
```

```julia
plot(sol2,plot_analytic=true)
```
